```python
class dog:
    
    species = "local"  #class attribute
    
    def __init__(self,name,age):
        
        #instance attribute
        self.name =name 
        self.age =age
    
    def speak(self,sound):
        self.sound = sound
        return f"{self.name} barks like {sound}"

class German(dog):

    # it is called method over ridding
        def speak(self , sound = "bow"):
            return f"{self.name} barks like {sound} blablabla"
            
class Indian(dog):
    # pass will also give same answer without creating function speak()
#     pass
    def speak(self,sound):
        return super().speak(sound)

class Us(dog):
    pass

buddy = German("buddy", 4)
moti = Indian("moti", 5)
john = Us("john",3)

print(moti.speak("grr"))
print(buddy.speak())  # with default argument 
print(buddy.speak("bhow-bhow")) # with user defined argument 
```

```python
class man:
    
    def __init__(self):
        self.name = "aman"
        self.age = 23
    
m1 = man()
print(id(m1))  # prints the location of object
```

```python
class man:
    
    def __init__(self):
        self.name = "aman"
        self.age = 23
    
    def update(self):
        self.age = 30
    
    def compare(self, other):
        if self.age == other.age:
            print("they are same")
        else:
            print("they are not same")
        
m1 = man()
print(id(m1))  # prints the location of object
print(m1.name)
print(m1.age)

m1.update()   # it updates age for m1 object
print(m1.age)

m2 = man()  
print("age of m2", m2.age)

m1.compare(m2)  # it compares m1 as self and m2 as other in compare function
```

# Types of method

```python
class student:
    
    def __init__(self,m1,m2,m3): #instance method
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
    
    def avg(self): #instance method
        return (self.m1+self.m2+self.m3)/3
    
    @classmethod
    def school(cls):
        return cls.school
    
    @staticmethod
    def info():
        print("name of the school")
    
        

s1 = student(45,86,95)
s2 = student(0,0,0)

print(s1.avg())
print(s2.avg())

print(student.school())
student.info()
```

```python
class a:
    def __init__(self):
        print("f1")
    
    def feature2(self):
        print("f2 is working")

class b(a): # it is still child class
    def __init__(self):
        super().__init__() # special way to call super class method
        print("f3")
        super().feature2() 
    
    def feature4(self):
        print("f4")

a1 = b()
```

# Polymorphism

```python
# Write Python3 code here

class car():

    # init method or constructor
    def __init__(self, node, color):
        self.node = node
        self.color = color

    def show(hfg):
        print("Model is", hfg.node )
        print("color is", hfg.color )

# both objects have different self which
# contain their attributes
audi = car("audi a4", "blue")
ferrari = car("ferrari 488", "green")

audi.show() # same output as car.show(audi)
ferrari.show() # same output as car.show(ferrari)

#note:we can also do like this
print("Model for audi is ",audi.node)
print("Colour for ferrari is ",ferrari.color)
#this happens because after assigning in the constructor the attributes are linked to that particular object
#here attributes(model,colour) are linked to objects(audi,ferrari) as we initialize them
# Behind the scene, in every instance method
# call, python sends the instances also with
# that method call like car.show(audi)
```

# Encapsulation

```python
class Enc:
    
    def __init__(self, name , salary):
        
        # public data member
        self.name =name
        # private data member
        self.__salary = salary
        
    
e = Enc("haris",34000)
print("name:",e.name)

# here we can't access private member
print("salary",e.__salary)
    
```

We can access private members from outside of a class using the following two approaches

Create public method to access private members

Use name mangling

```python
class Enc:
    
    def __init__(self, name , salary):
        
        # public data member
        self.name =name
        # private data member
        self.__salary = salary
        
    # making public method    
    def show(self):
        # private members are accessible from a class
        print("salary",self.__salary)
        
    
e = Enc("haris",34000)
print("name:",e.name)

# calling public method of the class
e.show()
```

```python
class Enc:
    
    def __init__(self, name , salary):
        
        # public data member
        self.name =name
        # private data member
        self.__salary = salary
        
    
e = Enc("haris",34000)
print("name:",e.name)

# direct access to private member using name mangling
print("salary",e._Enc__salary)
    
```

```python
# base class
class Company:
    def __init__(self):
        # Protected member
        self._project = "NLP"

# child class
class Employee(Company):
    def __init__(self, name):
        self.name = name
        Company.__init__(self)
#         or use this
#         super().__init__()

    def show(self):
        print("Employee name :", self.name)
        # Accessing protected member in child class
        print("Working on project :", self._project)

c = Employee("Jassi")
c.show()

# Direct access protected data member
print('Project:', c._project)
```

# Abstraction

```python
class car:
    def milege(self):
        print("hi")
        pass

class tata(car):
    def milege(self):
        print("tata 36")
class mahindra(car):
    def milege(self):
        print("mahindra 36")
class tesla(car):
    def milege(self):
        print("tesla 36")
        
# Here we can create object of parent class because it is not an abstract class
c = car()
c.milege()

t = tata()
t.milege()

m = mahindra()
m.milege()

tl = tesla()
tl.milege()
```

```python
from abc import ABC,abstractmethod

class bike(ABC):
    def __init__(self):
        print("it's a racing bike")
    
    @abstractmethod
    def speed(self):
        pass

class duke(bike):
    def speed(self):
        print("duke has speed of 200 kmph")

class bmw(bike):
    def speed(self):
        print("bmw has speed of 220 kmph")
    
class harly(bike):
    def speed(self):
        print("harly has speed of 250 kmph")    

# we can't create object of abstract class, if we do ,we will get error
# tw = bike()
# tw.speed()

d= duke()
d.speed()

b = bmw()
b.speed()

h = harly()
h.speed()
```
