```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn import tree
from sklearn.metrics import r2_score,classification_report,accuracy_score
from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing()
```

```python
dir(df)
```

```python
df.data.shape
```

```python
X = pd.DataFrame(data = df.data,columns=df.feature_names)
```

```python
X.head()
```

```python
y = pd.DataFrame(df.target)
```

```python
y.head()
```

```python
dtr_model = tree.DecisionTreeRegressor() 
```

```python
parameter = {
    "criterion" : ["squared_error", "friedman_mse", "absolute_error","poisson"],
    "splitter" : ["best", "random"],
    "max_depth": [1,2,3],
    "max_features" : ["auto", "sqrt", "log2"] 
}
```

```python
dtr_model_cv = GridSearchCV(dtr_model,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
dtr_model_cv.fit(X_train,y_train)
```

```python
dtr_model.fit(X_train,y_train)
```

```python
dtr_model_cv.best_params_
```

```python
y_pred = dtr_model_cv.predict(X_test)
```

```python
# score of hyper parameter tunnig and cross validation
r2_score(y_test,y_pred)
```
