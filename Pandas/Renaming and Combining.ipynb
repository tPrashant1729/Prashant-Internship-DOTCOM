```python
df1 = pd.DataFrame({
    "subject":["math","science","guj"],
    "marks":[34,56,78]
})

df2 = pd.DataFrame({
    "subject":["math","science","guj"],
    "marks":[72,45,98]

})
```

```python
df3=pd.merge(df1,df2,on="subject")
df3
```

```python
df1 = pd.DataFrame({
    "subject":["math","science","guj","hindi","s.s"],
    "marks":[34,56,78,45,39]
})

df2 = pd.DataFrame({
    "subject":["math","science","guj","biology"],
    "marks":[72,45,98,100]

})

# it is for intersaction
pd.merge(df1,df2,on = "subject",how="inner")
```

```python
# it is for union
pd.merge(df1,df2,on = "subject",how="outer")
```

```python
# it is for left union as order of dataframe
pd.merge(df1,df2,on = "subject",how="left")
```

```python
pd.merge(df1,df2,on = "subject",how="right")
```
