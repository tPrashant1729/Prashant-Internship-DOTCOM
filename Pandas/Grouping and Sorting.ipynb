```python
import pandas as pd
```

```python
mov = pd.read_csv("imdb-top-1000.csv")
```

```python
df = pd.DataFrame(mov)
```

```python
df.value_counts("Genre")
```

We can replicate what value_counts() does by doing the following:

```python
df.groupby("Genre").Genre.count()
```

```python
df.head(3)
```

```python
df.groupby("Genre").head(10)
```

```python
v=df.groupby("Genre")
v
for Genre,Genre_df in v:
    print(Genre)
    print(Genre_df)
```

```python
v.get_group("Comedy")
```

```python
df.groupby("Genre")[["IMDB_Rating","Series_Title"]].max()
```

```python
df.groupby('Genre').apply(lambda df: df.Series_Title.iloc[0])
```

```python
df.groupby(['Genre','Released_Year']).apply(lambda df: df.Series_Title.iloc[0])
```

```python
df.groupby(['Genre']).apply(lambda df: df.loc[df.IMDB_Rating.idxmax()])
```

```python
# Gross.idxmax() will return index with highest gross
df.loc[df.Gross.idxmax()]
```

```python
v=df.groupby("Released_Year")
v.get_group("2015")
```

```python
df
```

```python
# To apply multiple function
df.groupby("Released_Year").Gross.agg([len,max,min])
```

```python
df.groupby(["Genre"]).Gross.agg([min,max,len])
```

```python
df.groupby(["Genre","IMDB_Rating"]).agg([min,max,len])
```

```python
v=df.groupby(["Genre","IMDB_Rating"]).Gross.agg([min,max,len])
mi = v.index
type(mi)
```

```python
v.reset_index()
```

## Sorting 

```python
# dorting works on only grouped by one parameter
v=v.reset_index()
v.sort_values(by="len")
```

```python
v.sort_index()
```

```python
v.sort_values(by=["IMDB_Rating","len"],ascending = False)
```
